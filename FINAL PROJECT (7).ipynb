</h3>
</div>

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Select required columns and remove missing values
df_3d = df[[
    "MORT Group Measure Count",
    "Safety Group Measure Count",
    "READM Group Measure Count"
]].dropna()

fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(
    df_3d["MORT Group Measure Count"],
    df_3d["Safety Group Measure Count"],
    df_3d["READM Group Measure Count"],
    alpha=0.7
)

ax.set_xlabel("Mortality")
ax.set_ylabel("Safety")
ax.set_zlabel("Readmission")

plt.title("3D Scatter Plot")
plt.show()
```
